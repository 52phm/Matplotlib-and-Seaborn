### Example 1

学习知识点：

1、x 轴刻度线标签

2、设置标题、x 轴和 y 轴的标签

### 1、准备数据

```python
import pandas as pd
data = pd.read_csv('../data/unrate.csv')
data.head()
```

```python
# 判断是否有空值
nan = data.isnull()
nanNum = nan.sum()
nanNum
```

### 2、创建空图形

```python
import matplotlib.pyplot as plt
plt.plot()
plt.show()
```

### 3、自定义绘图

```python
# 选取 [0: 12] 区间的数据画图
plotData = data[0: 12]
plt.plot(plotData.DATE, plotData['VALUE'])
plt.show()
```

知识点1：X 轴刻度线标签处理

```python
# 针对上面的横坐标重叠问题进行优化
plt.plot(plotData.DATE, plotData['VALUE'])
''' 
将 X 轴刻度线标签旋转 45(60,90等) 度，这样它们就
不会重叠,可以使用浮点或整数值指定旋转度。
'''
plt.xticks(rotation=45)
plt.show()
```

知识点2：设置图像标题、x 轴和 y 轴标签值

```python
plt.plot(plotData.DATE, plotData['VALUE'])
plt.xticks(rotation=90)
plt.title('Monthly Unemployment Trends, 1948')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.show()
```

### Example 2

学习知识点：

1、设置画布大小

2、设置图形颜色

3、设置图例

### 1、准备数据

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('../data/unrate.csv')
data.head()
```

```python
# 将 DATE 列设置为 datetime 时间索引列
data['DATE'] = pd.to_datetime(data['DATE'])
data['MONTH'] = data['DATE'].dt.month 
data.head()
```

### 2、创建空图形

知识点1：设置画布大小

plt.figure(figsize=(6, 3))

```python
''' 设置画布大小 '''
plt.figure(figsize=(6, 3))
plt.plot()
plt.show()
```

### 3、自定义绘图

知识点2：设置颜色

plt.plot(x, y, color='b')

plt.plot(x, y, c='b')

color='b' 等效于 c='b'

```python
plt.plot(data[: 12]['MONTH'], data[: 12]['VALUE'], color='b')
plt.plot(data[12: 24].MONTH, data[12: 24].VALUE, c='r')
plt.show()
```

```python
知识点3：图例设置

第一：设置图例标签名，plt.plot(label='图例标签名')

第二：设置图例放置的位置，plt.legend(loc='放置的位置')

loc 的参数有：

        ===============   =============
        Location String   Location Code
        ===============   =============
        'best'            0
        'upper right'     1
        'upper left'      2
        'lower left'      3
        'lower right'     4
        'right'           5
        'center left'     6
        'center right'    7
        'lower center'    8
        'upper center'    9
        'center'          10
        ===============   =============
```

```python
# 批量设置图形颜色
plt.figure(figsize=(10, 7))
color = ['red', 'blue', 'green', 'orange', 'black']
for i in range(5):
    start_index = i * 12
    end_index = (i + 1) * 12
    ''' 设置图例标签名 '''
    label = str(1948 + i)
    x_plot = data[start_index: end_index]
    plt.plot(x_plot['MONTH'], x_plot['VALUE'], color=color[i], label=label)
plt.legend(loc='best')
plt.xlabel('Month--')
plt.ylabel('Nurate--')
plt.title('Monthly Unemployment Trends, 1948-1952')
plt.show()
```
