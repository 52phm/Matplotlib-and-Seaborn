```python
# ----------------------------------
# Example 1
import pandas as pd
import matplotlib.pyplot as plt

womenDegrees = pd.read_csv('../data/womendegrees.csv')
# Year 年份、Biology 生物学
fig, ax = plt.subplots()
plt.plot(womenDegrees['Year'], womenDegrees['Biology'])
plt.title('Biology')
plt.xlim(1970, 2000)
plt.legend(loc='best')
plt.show()
```

```python
# ----------------------------------
# Example 2
# 100-女性学位意味着男性
plt.plot(womenDegrees['Year'], womenDegrees['Biology'], label='women')
plt.plot(womenDegrees['Year'], 100-womenDegrees['Biology'], label='men')
# 通过rcParams设置全局横纵轴字体的大小
# plt.rcParams['xtick.labelsize'] = 24
# plt.rcParams['ytick.labelsize'] = 24
# 设置字体样式，这里设置为中文黑体
plt.rcParams['font.family'] = 'SimHei'
plt.legend(loc='best')
plt.title('生物学博士')
plt.show()
```

```python
# ----------------------------------
# Example 3
# 100-女性学位意味着男性
fig, ax = plt.subplots()
ax.plot(womenDegrees['Year'], womenDegrees['Biology'], label='women')
ax.plot(womenDegrees['Year'], 100-womenDegrees['Biology'], label='men')

# 隐藏刻度线
ax.tick_params(bottom='off', top='off', left='off', right='off')
# 隐藏坐标轴线
for key,spine in ax.spines.items():
    spine.set_visible(False)
    
ax.legend(loc='best')
ax.set_title('Biology')
plt.show()
```

```python
# ----------------------------------
# Example 4
fig = plt.figure()
data = womenDegrees

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(data['Year'], data['Computer Science'], color='r', label='women',
         linewidth=5)
ax1.plot(data['Year'], 100-data['Computer Science'], color='g', label='men', 
         marker='>', markerfacecolor='r', markersize=12)

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(data['Year'], data['Education'], label='women', linestyle='-.')
ax2.plot(data['Year'], 100-data['Psychology'], label='men')

ax1.legend(loc='best')
ax2.legend(loc='best')
plt.subplots_adjust(left=1, right=3, top=1.5, bottom=1)
plt.show()
```

```python
# ----------------------------------
# Example 5
# 100-女性学位意味着男性
fig = plt.figure()
item = ['Computer Science', 'Education', 'Psychology', 'Physical Sciences']
for num in range(0, 4):
    ax = fig.add_subplot(2, 2, num + 1)
    ax.plot(womenDegrees['Year'], womenDegrees[item[num]], color='r', label='women',linewidth=5)
    ax.plot(womenDegrees['Year'], 100-womenDegrees[item[num]], color='g', label='men', marker='>')
    ax.legend(loc='best')
    ax.margins(x=0.2, y=0.3)
plt.subplots_adjust(left=1, right=3, top=3, bottom=1)
plt.show()
```

```python
data = womenDegrees
fig = plt.figure()
fig.subplots_adjust(left=0.7, right=2.5, top=1.3, 
                    bottom=0.4, hspace=1.5, wspace=0.7)
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.plot(data['Year'], data['Education'], c='k')
ax1.plot(data['Year'], 100-data['Education'], c='g')
ax2.plot(data['Year'], data['Psychology'])
ax2.plot(data['Year'], 100-data['Psychology'])

plt.rcParams['font.family'] = 'SimHei'

''' 图像文本和标注'''
ax1.text(1980, 78, '女性', fontsize=20)
ax1.text(1980, 25, '男性', fontsize=20)

ax2.annotate(s='lxz', xy=(1980, 65.1), xytext=(1970, 74.1), xycoords='data',textcoords='data', 
             arrowprops=dict(arrowstyle="->",connectionstyle='arc3',color='r',), 
             fontsize=24, color='g')
plt.show()
```
