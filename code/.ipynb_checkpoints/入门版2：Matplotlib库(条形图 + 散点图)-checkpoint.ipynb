### 增加子图

```python
import matplotlib.pyplot as plt

''' 创建空图形 '''
# 画布(两种方法)
fig = plt.figure()         # 这种方式产生的子图图形之间会分隔开，效果比较好
fig, ax = plt.subplots()   # 这种方式产生的 fig 增加子图坐标轴会重叠
# 在画布上添加 4 个子图
ax1 = fig.add_subplot(3, 2, 1)
ax2 = fig.add_subplot(3, 2, 2)
ax3 = fig.add_subplot(3, 2, 4)
ax4 = fig.add_subplot(3, 2, 6)
plt.show()
```

```python
import numpy as np

''' 第一步：准备数据 '''
x1, y1 = np.random.randint(1, 5, 5), np.arange(5)
x2, y2 = np.arange(10) * 3, np.arange(10)
''' 第二步：创建空图形 '''
fig = plt.figure()
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(2,1, 2)
''' 第三步：自定义绘图 '''
ax1.plot(x1, y1)
ax2.plot(x2, y2)
''' 第四步：保存画布 '''
plt.savefig('../tmp/subplot.png')  # 常规保存
plt.savefig('../tmp/subplot1.png', transparent=True) # 保存透明画布
''' 第五步：显示图形'''
plt.show()
```

### 3、自定义绘图

## Example 1  条形图

```python
''' 数据 '''
cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
data_1 = data[cols]
bar_positions = np.arange(1,6) 
bar_heights = data_1.loc[0, cols].values

import matplotlib.pyplot as plt
fig, ax = plt.subplots()
''' 自定义绘图 '''
ax.bar(bar_positions, bar_heights, 0.5)
plt.show()
```

```python
''' 数据 '''
cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
data_1 = data[cols]
bar_positions = np.arange(1,6)
bar_heights = data_1.loc[0, cols].values

import matplotlib.pyplot as plt
fig, ax = plt.subplots()
''' 自定义绘图 '''
ax.bar(bar_positions, bar_heights, 0.5)

# ---------------------------------
# 自定义设置 X 轴刻度和对应的标签
tick_positions = range(1,6)
# ax.set_xticks(tick_positions)
ax.set_xticklabels(cols, rotation=78)

# --------------------------------
# 设置 X 轴、Y 轴标签和标题
# ax.set_xlabel('Rating Source')
# ax.set_ylabel('Average Rating')
# ax.set_title('2015')
# ax.set_xlim(0, 6)
# ax.set_ylim(0, 6)
ax.set(title='2015',             # 标题
       xlabel='Rating Source',   # X 轴标签 
       ylabel='Average Rating',  # Y 轴标签
       xticks=tick_positions,    # X、Y 轴刻度
       xlim=[0, 6], ylim=[0, 6], # X、Y 轴范围限制
      )
plt.show()
```

```python
''' 数据 '''
cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
data_1 = data[cols]
bar_positions = np.arange(1,6) 
bar_heights = data_1.loc[0, cols].values

import matplotlib.pyplot as plt
fig, ax = plt.subplots()
''' 自定义绘图 '''
ax.barh(bar_positions, bar_heights, 0.5)

# ---------------------------------
tick_positions = range(1,6)
# 没有 rotation=45 参数才可以统一到 ax.set()中
# ax.set_yticklabels(cols, rotation=45)


ax.set(title='2015',             # 标题
       xlabel='Rating Source',   # X 轴标签 
       ylabel='Average Rating',  # Y 轴标签
       yticks=tick_positions,    # X、Y 轴刻度
       xlim=[0, 6], ylim=[0, 6], # X、Y 轴范围限制
       yticklabels=(cols)        # 设置 Y 轴刻度标签
      )
plt.show()
```

## Example 2 散点图

```python
import matplotlib.pyplot as plt
cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
data_2 = data[cols]
fig, ax = plt.subplots()
ax.scatter(data_2['RT_user_norm'], data_2['Metacritic_user_nom'], )
ax.set(xlabel='RT_user_norm',
        ylabel='Metacritic_user_nom'
       )
plt.show()
```

```python
import matplotlib.pyplot as plt
cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Stars']
data_2 = data[cols]

fig = plt.figure(figsize=(5, 10))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)
ax1.scatter(data_2['Fandango_Stars'], data_2['RT_user_norm'], color='r', )
ax1.scatter(data_2['Metacritic_user_nom'], data_2['IMDB_norm'], color='black')
ax2.scatter(data_2['IMDB_norm'], data_2['Fandango_Stars'], color='green')

ax2.set(xlabel='IMDB_norm',
        ylabel='Fandango_Stars',
       )
plt.show()
```
