```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
''' 
设置 %matplotlib inline 和不设置的区别：
1、设置，就不用使用 plt.show() 就可以展示图像
2、不设置，要展示图像，必须使用 plt.show() 才能展示
'''
%matplotlib inline
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sinplot()
```

```python
# 设置或重置 Seaborn 默认值
sns.set()  
sinplot()
```

```python
# --------------------------------
知识点1：Seaborn 样式
''' 设置或重置 seaborn 默认值 '''
sns.set()  

''' 设置画布主题风格 '''
sns.set_style(style=None, rc=None)

参数介绍
--------
style: 有五种风格
    darkgrid, whitegrid, dark, white, ticks
rc: dict
    
Examples
--------
>>> set_style("whitegrid")
>>> set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})
    
''' 用with设置临时样式 '''
sns.axes_style('whitegrid')

例子
--------
>>>with sns.axes_style('white'):
        sns.boxplot()
```

```python
sns.axes_style('dark')
sinplot()
```

```python
sns.set_style("whitegrid")
# 生成 20 行 6 列的二维样本数据
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

```python
sns.set_style('ticks')
sinplot()
```

```python
sinplot()
# 坐标轴线设置
sns.despine(offset=10)
```

```python
# 小提琴图
sns.violinplot(x=data)
```

```python
# 用设置临时样式
with sns.axes_style('darkgrid'):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sns.set_style('dark')
sinplot(-1)
```

```python
# ---------------------------------
知识点2：上下文函数
sns.set_context(context=None, font_scale=1, rc=None)

文档介绍
-----------
用来联系上面设置过的图形样式、风格，相当于继承上面的功能

参数介绍
-----------
context: 用来设置怎样的画布效果
    paper, notebook, talk, poster
    
例子
--------
>>> set_context("paper")

>>> set_context("talk", font_scale=1.4)

>>> set_context("talk", rc={"lines.linewidth": 2})
```

```python
sns.set_context('paper')
plt.figure(figsize=(8, 6))
sinplot()
```

```python
sns.set_context('notebook')
plt.figure(figsize=(8, 6))
sinplot()
```

```python
sns.set_context('talk', font_scale=1.5)
plt.figure(figsize=(8, 6))
sinplot()
```

```python
sns.set_context('poster', font_scale=1.5, rc={'lines.linewidth': 5})
plt.figure(figsize=(8, 6))
sinplot()
```
