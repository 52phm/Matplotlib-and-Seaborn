```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('../data/fandango_scores.csv')

''' 数据 '''
x1 = data['Fandango_Ratingvalue']
x2 = data['RT_user_norm']
x3 = data['Metacritic_user_nom']
x4 = data['IMDB_norm']

''' 子图画布 '''
fig = plt.figure()
# 子图间距调整
fig.subplots_adjust(wspace=0.5,
                    hspace=0.3, 
                    left= 0.1,
                    right=2,
                    top=2,
                    bottom=0.5
                   )
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

''' 子图直方图 '''
ax1.hist(x1, bins=20, range=(0, 5))
ax2.hist(x2, bins=20, range=(0, 5))
ax3.hist(x3, bins=20, range=(0, 5))
ax4.hist(x4, bins=20, range=(0, 5))

''' 设置直方图的标题、y 轴限定范围 '''
ax1.set(title='Distribution of Fandango Ratings', ylim=(0, 50))
ax2.set(title='Distribution of Rotten Tomatoes Ratings', ylim=(0, 50))
ax3.set(title='Distribution of Metacritic Ratings', ylim=(0, 50))
ax4.set(title='Distribution of IMDB Ratings', ylim=(0, 50))

''' 设置坐标轴边线 '''
# 隐藏坐标轴线
ax1.spines['right'].set_visible(False)
# 设置底部边线的位置为 outward
ax2.spines['bottom'].set_position(('outward', 15))
# 隐藏刻度线
ax3.tick_params(bottom='off', top='off', left='off', right='off')
# 隐藏刻度线、坐标轴线、刻度值
ax4.axis('off')
# 设置内边距
ax3.margins(x=0.2, y=0.3)
plt.show()
```

### Example 2(箱形图)

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('../data/fandango_scores.csv')
x = data['IMDB_user_vote_count']

fig, ax = plt.subplots()
ax.boxplot(x)
ax.set_ylim((0, 140000))
plt.show()
''' 
离群点：这里有两个圆圈析出，代表是离群点，
离散度：盒子的长度集中在 3000~40000 之间
'''
```

```python
x1 = data['RT_norm_round']
x2 = data['RT_user_norm_round']
x3 = data['Metacritic_user_norm_round']
x4 = data['IMDB_norm_round']

fig, ax = plt.subplots()
ax.boxplot([x1, x2, x3, x4])

# 设置 x 刻度线的标签
cols = ['RT_norm_round', 'RT_user_norm_round', 
        'Metacritic_user_norm_round', 'IMDB_norm_round']
ax.set_xticklabels(cols, rotation=90)
plt.show()
```
